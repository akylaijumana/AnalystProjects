```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/Sample - Superstore.csv', encoding ='latin1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#checking nulls
df.isnull().sum()
```

```python
#rename columns (replacing underscores, lowercase)
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_').str.replace('-','_')
df.columns.tolist()
```

```python
df.dtypes
```

```python
#fix date columns
df['order_date']=pd.to_datetime(df['order_date'])
df['ship_date']=pd.to_datetime(df['ship_date'])
df.dtypes
```

```python
#remove duplicates
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df[['sales','profit','discount','quantity']].describe()
```

```python
#Days to ship
df['days_to_ship']= (df['ship_date']-df['order_date']).dt.days
```

```python
#profit margin %
df['profit_margin'] = round((df['profit']/df['sales'])*100,2)
df.head()
```

```python
from google.colab import files
df.to_csv('superstore_cleaned.csv', index=False)
files.download('superstore_cleaned.csv')
```
